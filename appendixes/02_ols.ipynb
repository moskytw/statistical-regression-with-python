```python
df_fair = sm.datasets.fair.load_pandas().data
df_fair_sample = df_fair.sample(
    frac=0.1, random_state=20190425
)
```

```python
df = df_fair_sample
sns.regplot(data=df, x='rate_marriage', y='affairs',
            x_jitter=1/2, y_jitter=20/2)
```
